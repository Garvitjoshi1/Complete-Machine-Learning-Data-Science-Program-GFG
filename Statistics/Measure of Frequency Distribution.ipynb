## Relative Frequency

````python
cars_in_country = pd.crosstab(df['origin'], "Number of cars") #type:ignore
````

```Adding a new table name relative frequency```

````python
cars_in_country['Relative Frequency'] = cars_in_country['Number of cars']/cars_in_country['Number of cars'].sum()
````

``` Round off value to 2 decimal place```

````python
cars_in_country['Relative Frequency'] = round(cars_in_country['Relative Frequency'], 2)
````

````python
cars_in_country
````

## Percentage Frequency
``` Relative Frequency × 100 = Percentage Frequency```

````python
cars_in_country['Percentage Frequency'] = cars_in_country['Relative Frequency'] * 100
````

````python
cars_in_country
````

```To get the percetage in pie chart use plot(kind = "pie", autopct = "%1.f%%")```

````python
df['origin'].value_counts().plot(kind = "pie", autopct = "%1.f%%")
````
