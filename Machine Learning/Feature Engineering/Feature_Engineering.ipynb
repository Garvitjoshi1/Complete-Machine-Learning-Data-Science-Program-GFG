#### Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#### Loading the data

```python
data = pd.read_csv("global_heat_index.csv")
data.head(10)
```

```python
data.info()
```

### <font color = "green">Correlation Matrix</font>

#### Cleaning the data:

```python
clean1 = data[data['Hour'] < 6]
clean2 = data[data['Hour'] > 18]

print("Cleaning data from 12:00am to 5:59am")
clean1.head(10)
```

```python
print("Cleaning data from 18:00 pm to 12:00 am")
clean2.head(10)
```

```python
data = data.drop(clean1.index, axis = 0)
data = data.drop(clean2.index, axis = 0)
```

```python
data.shape
```

```python
data.describe()
```

### Input and Output Data(separation)
``Min-Max Scaling``

```python
x = data.iloc[:,0:10]
y = data.iloc[:, -1]  
```

```python
from sklearn.preprocessing import MinMaxScaler

x = pd.DataFrame(MinMaxScaler().fit(x).transform(x))
```

```python
x.head()
```

```python
data.head()
```

### Correlation within the Columns:

```python
corr_matrix = data.corr()
```

```python
corr_matrix
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(data.corr(), annot=True)
```

### <font color ="green">Extra_Tree_Classifier</font>

```python
from sklearn.ensemble import ExtraTreesClassifier
model = ExtraTreesClassifier()
model.fit(x, y)
```

```python
col = ["Year", "Month", "Day", "Hour", "Dew Point", "Temperature", "Pressure", "Relative Humidity", "Wind Direction", "Wind Speed"]
feature_imp = pd.Series(model.feature_importances_, index = col)
feature_imp
```

```python
feature_imp.nlargest(10).plot(kind = "bar")
plt.show()
```

#### <font color = "green">Select_K_Best_Method</font>

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
bestfeatures = SelectKBest(score_func=chi2, k = 10).fit(x, y)
```

```python
scores = pd.DataFrame(bestfeatures.scores_)
columns  =pd.DataFrame(col)
```

```python
feature_score = pd.DataFrame(bestfeatures.scores_)
columns = pd.DataFrame(col)
feature_score = pd.concat([columns, scores], axis = 1)
feature_score.columns = ['features', 'Scores']
```

```python
feature_score
```

```python
feature_score.nlargest(10, "Scores").plot(kind = "bar")
plt.show()
```
