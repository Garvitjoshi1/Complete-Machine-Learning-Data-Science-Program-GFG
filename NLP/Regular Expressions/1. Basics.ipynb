## 1.$Search$ only for first occurence
`stores value only if the value is present in another string.`

```python
srch = re.search("you", "Hello, How are you?")
srch
```

`span=(15, 18) above is the index of where the value lies in string.`

```python
srch.span()
```

```python
srch.start()
```

```python
srch.end()
```

```python
a = "Hello, How are you?"
b = "are"
re.search(b, a)
```

## 2. $findall()$
`frequency count of number of times a string occur in another.`

```python
a = "Hello, How are area in your house ?"
b = "are"
re.findall(b, a)
```

```python
a = "hello"
b = "why"
re.findall(b, a)
```

## 3. $finditer(): $ to find all spans of occurence in string

```python
for s in re.finditer("are", "Hello, How are area you?"):
    print(s.span())
```

## $4.Creating$ $ Regular$  $Expression: $

```python
txt = "My phone number is 753483947"

pattern = "\d\d\d\d\d\d\d"
re.search(pattern, txt)
```

```python
txt = "My phone number is 753-483-947"

pattern = "\d\d\d-\d\d\d-\d\d"
re.search(pattern, txt)
```

```python
# above code can also be written as:
txt = "My phone number is 753-483-947"

pattern = "\d{3}-\d{3}-\d{1}"
re.search(pattern, txt)
```

```python
txt = "My phone number is 753-483-947"

pattern = '\w\w\w\w\w\w'
re.search(pattern, txt)
```

```python
re.search(pattern, txt).group()
# to get the value printed
```

```python
re.search("\d\d", txt).group()
# to get the value printed in customed pattern also
```
