```python
import re
```

### 1. Finding pattern with specific ending

```python
re.findall('at', 'the rat sat on the mat where there were cats')
```

```python
# add fullstop(.) at starting of pattern to make it stop by following pattern
re.findall('.at', 'the rat sat on the mat where there were cats')
```

### 2. Start and End Pattern Matching

```python
re.findall("\d", "4 is divisible by 2 not by 3")
```

```python
# find number at last
re.findall("\d$", "4 is divisible by 2 not by 3")
```

```python
# find doubl-digit number at last
re.findall("\d\d$", "4 is divisible by 2 not by 32")
```

```python
# find digit number at beginning
re.findall("^\d", "4 is divisible by 2 not by 32")
```

### 3. Finding Capital Letters

```python
re.findall('[A-Z]', "Welcome to GFG")
```

### 4. Finding Small Letters

```python
re.findall('[a-z]', "Welcome to GFG")
```

### 5. Finding Upper and Lower Case

```python
re.findall('[A-Za-z]', "Welcome to GFG")
```

### 6. Finding UpperCase, LowerCse and Numbers:

```python
re.findall('[A-Za-z0-9]', "Welcome to GFG121")
```
