```python
import nltk

nltk.download('stopwords')

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
txt = "This is not a good time to talk"
```

```python
stopwords.words()
```

```python
stopwords.words('english')
```

```python
print(word_tokenize(txt))
```

```python
stopword = stopwords.words('english')
```

```python
txt = word_tokenize(txt)
```

```python
# to identify if the word is stop_word or not.
for word in txt:
    print(word, word in stopword)
```

``Lower-Casing the complete sentence to fetch all the stop words``

```python
txt = "This is not a good time to talk"
```

```python
txt = txt.lower()
```

```python
txt
```

```python
txt = word_tokenize(txt)
```

```python
for word in txt:
    print(word, word in stopword)
```

`` to find words in sentence that are not stopwords``

```python
for word in txt:
    if(word.lower() not in stopword):
        print(word)
```
